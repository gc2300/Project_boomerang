# Define the metrics
We need to decide how we are going to measure whether the emails are successful.
1. Overall business objective: increase the number of orders made using our site by bringing back the registered users.
    * To reach the objective, the product team launched a new feature and collaborated with the marketing team to start an email campagin to inform registered users about the new feature. When users click on the link inside the emails, they will be brought to the company's website.
    * We are working with the marketing team to evaluate the impact of the current email campaign and try to improve the probability that users click on the link inside the emails.
2. Let's use a **email funnel analysis** to break the objective down:
    * The email was not delivered in the inbox (ended up in spam folder)
    * The email was delivered in the inbox: 
        * The user did not check inbox.
        * Users saw the email in the inboox:
            * Users saw the email title in the inbox but did not click on it
            * **Users opened the email**:
                * Users did not reach to the part where the link was shown
                * Users saw the link:
                    * Users did not click
                    * Users clicked
                        * Users clicked by accident
                        * **Users clicked and statyed on the new feature page on our website for at least 5 seconds.**
                            * Users abandoned website without make a purchase
                            * **Users made a purchase**
3. We can pick the most important steps in the above and define the final funnel.
    * Email delivered in the inbox
        * Email opened
            * Link clicked and checked some pages on our website
                * Made a purchase 
4. #### Select metric for each step. If it's a key point, let's use a count metric. If it's between key points, let's use a rate or probability.
    * Number of emails delivered in the inbox
        * Number of emails opened within 7 days after sent
            * **Click-through-probability of the link**: the number of users clicked on the link and stayed on the new feature page for at least 5 seconds divided by the number of users who opened the email. This will be the **main metric** we try to **optimize**.
                * **Conversion rate**: the number of users clicked the link and made at least one purchase on the same day divided by the number of users opened the email. This will be the **counter metrics** to ensure the optimization of the main metric is leading to a sustainable growth of the product.

### Outliers handling: Create an outlier free training set:

```python
def remove_outlier(df, column):
    df_copy = df.copy()
    Q1 = df_copy[column].quantile(0.25)
    Q3 = df_copy[column].quantile(0.75)
    IQR = Q3 - Q1
    Lower_Fence = Q1 - (1.5 * IQR)
    Upper_Fence = Q3 + (1.5 * IQR)
    return df_copy[(df_copy[column] > Lower_Fence) & (df_copy[column] <Upper_Fence)]
```

```python
clean_train_df_p1 = remove_outlier(train_df_p1, 'hour')
```

```python
clean_train_df_p1 = remove_outlier(train_df_p1, 'user_past_purchases')
```

### Binning numerical variable
Here we bin the numerical variables but also keep the original variables in the dataframe

```python
def binning(df):
    df_copy = df.copy()
    #Bin the variables
    bins_value = [0] + [np.quantile(df_copy['user_past_purchases'], 0 + x*0.2) for x in range(6) if x>0]
    bins_tag = []
    for i in range(len(bins_value)):
        if i == 0:
            bins_tag.append('[%d,%d)'%(bins_value[i], bins_value[i+1]))
        if (i > 1) and (i<len(bins_value)-1):
            bins_tag.append('[%d,%d)'%(bins_value[i-1], bins_value[i]))
        if i == len(bins_value)-1:
            bins_tag.append('[%d,%d]'%(bins_value[i-1], bins_value[i]))
    df_copy['user_past_purchases_binned'] = pd.cut(df_copy['user_past_purchases'], bins=bins_value, include_lowest=True, labels=bins_tag)
    
    #Hour
    df_copy['hour_binned'] = pd.cut(df_copy['hour'], bins=[1,5, 13, 21, 24], include_lowest=True, labels=['night', 'morning', 'afternoon', 'night2'])
    #replace night2 with night
    df_copy['hour_binned'] = df_copy['hour_binned'].replace('night2', 'night').cat.remove_unused_categories()
    return df_copy
```

```python
def numerical_normalize(df, train_df):
    """
    df is the dataframe to be normalized
    train_df is the dataframe that the normalization scaler fit from. 
    We want both training and test set use the same normalization scaler, which is fitted from training set.
    """
    df_copy = df.copy()
    #Select numerical columns
    numerical_cols = df_copy.select_dtypes(['int64', 'float64']).columns.tolist()
    print( "These columns will be normalized: ", ", ".join(numerical_cols)  )
    scaler = MinMaxScaler()
    scaler.fit(train_df[numerical_cols])
    df_copy.loc[:,numerical_cols] = scaler.transform(df_copy[numerical_cols])
    return df_copy
```

```python
#normalize x_train_p1
norm_clean_X_train_p1 = numerical_normalize(clean_X_train_p1, clean_X_train_p1)
```

```python
def cate_encoding(df, drop_first):
    df_copy = df.copy()
    #Before building the regression, we need to know which ones are the reference levels for the categorical variables
    #only keep categorical variables
    obj_cols = df_copy.select_dtypes(['object']).columns.tolist()
    df_copy.loc[:,obj_cols] = df_copy[obj_cols].astype('category')
    #only keep categorical variables
    df_categorical = df_copy.select_dtypes(['category'])
    #find reference level, i.e. the first level
    print(df_categorical.apply(lambda x: x.cat.categories[0]))
    df_dummy = pd.get_dummies(df_copy, drop_first=drop_first)
    return df_dummy
```

#### sample size of each combination of values of features
* In this email campaign, we covered all 840 combinations of 'email_version', 'hour_binned','weekday', 'user_country', and 'user_past_purchases_binned'. On average, we sent around 83 emails to each combination. The minimum is only 4 and the 25th quantile is 22, but we have a 75th quantile at 96 and max at 525! This means the emails were not sent in a randomized way to give each combination a similar number of emails.

#### Why we care about this:
* When we are evaluate average customer's interaction with email campaign:
    * i.e. using linear models like logistic regression
    * What an average customer is defined by the distribution of features in our analysis. We need to know what's an average/typical customer look like from our analysis. This requires us to see what are the segments of users that receive the most emails, which means they are more represented the average user depiction.
* When we are researching about speciifc segments of customer's interaction with email campaign:
    * i.e. using decision tree based methods

```python
def summarize_sample(df):
    sample_cols = ['email_version', 'weekday', 'user_country', 'user_past_purchases', 'hour']
    sample_cases = []
    for i in range(df.shape[0]):
        cur_list = []
        for j in range(df.shape[1]):
            if df.iloc[i, j] == 1: 
                cur_list.append(list(df.columns)[j])
        sample_cases.append(cur_list)
    sample_result_df = pd.DataFrame(sample_cases, columns = sample_cols, index = df.index )
    return sample_result_df.join(df['count']) #.sort_values(['email_version', 'weekday', 'user_country', 'user_past_purchases', 'hour'])
```

### Take a look at types of independent variables
All the independent variables have potential influence to the dependent variables. Some vairables that are manipulated by us and some variables are extra factor but not modifiable by an experiment. Let's categorize the variables so we have a better idea how to explore them:

* Manipulatable variables:
    * **email_id**: Unique id for each email
    * **email_version**: personalized emails have the name of the person receiving the email. generic emails have generic greeting.
    * **hour**: the local time on which the email was sent
    * **weekday**: the weekday on which the email was sent

* Extra factors:
    * Participant variables:
        * **user_id**: Unique id for each user
        * **user_country**: the country where the person receiving the mail is based. It comes from the user ip address when they registered the account.
        * **user_past_purchases**: how many times the user has purchased from the website after registration
